# Perfome meaningful visualization using the dataset.bring atleast 3 visualization

# Histogram

```python
plt.figure(figsize=(10,7))
plt.hist(data['SL'],color='blue',rwidth=0.8,density=True,cumulative=True)
plt.title("distribution of SW & SL")
plt.xlabel("SL")
plt.ylabel("SW")
```

# pair plots

```python
sns.pairplot(data)
data=data[['SL','SW','PL','PW']]
sns.pairplot(data)
```

# scatter plot

```python
plt.figure(figsize=(7,5))
plt.scatter(data["SL"],data['PL'],s=30,c='red')
plt.title('Plot of SL vs PL',fontsize=18)
plt.xlabel('SL')
plt.ylabel('PL')
```
